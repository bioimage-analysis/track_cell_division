```python
def clean_d2(liste_a):
    for i in range(len(liste_a)):
        for j in range(len(liste_a)):
            #print(j)
            try:
                ar_daughter2 = liste_a[j][2][:,0:2]
                ar_mom = liste_a[i][0][:,0:2]
                intersect = np.in1d(ar_daughter2, ar_mom)
                intersect_b = intersect.reshape(int(len(intersect)/2), 2)

                liste_a[j][2][:,0:2][intersect_b]=0                
            except IndexError:
                pass
    for k in range(len(liste_a)):
        try:
            ar = (liste_a[k][2][:,0:2])
            liste_a[k][2][:,0:2][np.where(ar==0)] = np.nan
        except IndexError:
            pass
    return liste_a
    
```

```python
#level the data because curvature at the edge:
def data_level(line):
    if line.ndim == 1:
        size = line.size
    else:
        size = line[0].size
    c, res, _, _, _ = np.polyfit(np.arange(0,size), np.mean(line, axis=0), 1, full=True)
    p = np.poly1d(c)
    x_new = np.linspace(np.arange(0,size)[0], np.arange(0,size)[-1], 20)
    y_new = p(x_new)
    line_leveled = np.mean(line, axis=0)/y_new
    result = np.sum(np.absolute(np.gradient(line_leveled)))/size
    return(line_leveled, result)
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,6))

line_flat, data = data_level(line)
print(data)
print(np.sum(np.absolute(np.gradient(np.mean(line, axis=0)))/line[0].size))
ax[0].plot(line_flat)
ax[0].set_title("flatten")
ax[1].plot(np.mean(line, axis=0))
ax[1].set_title("row")
```
